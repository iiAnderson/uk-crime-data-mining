```python
import pandas as pd
import numpy as np

crime = pd.read_csv("/home/robbie/Desktop/dat.csv")
crime.columns = ["Month", "Longditude", "Latitude", "LSOA", "Crime type", "Postcode", "Postcode2", "House Price", "Population", "Deprivation Index"]

from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()
crime["unique_LSOA"] = lb_make.fit_transform(crime["LSOA"])
crime["unique_Month"] = lb_make.fit_transform(crime["Month"])


crime.head()
```

```python
X = crime.iloc[:, [1, 2, 7, 8, 9]].values
Y = crime.iloc[:, 4].values
               
print(X)
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20)  
```

```python
from sklearn.preprocessing import StandardScaler  
scaler = StandardScaler()  
scaler.fit(X_train)

X_train = scaler.transform(X_train)  
X_test = scaler.transform(X_test)  
```

```python
from sklearn.neighbors import KNeighborsClassifier  
classifier = KNeighborsClassifier(n_neighbors=5)  
classifier.fit(X_train, y_train)  
y_pred = classifier.predict(X_test)  

from sklearn.metrics import classification_report, confusion_matrix  
from sklearn.model_selection import cross_val_score
    
print(confusion_matrix(y_test, y_pred))  
print(classification_report(y_test, y_pred))  
```

```python

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix  
from sklearn.model_selection import cross_val_score
from sklearn import tree

clf_gini = DecisionTreeClassifier(criterion = "gini", random_state = 100)
clrTree = clf_gini.fit(X_train, y_train)
outTree = clrTree.predict(X_test)

print(confusion_matrix(y_test, outTree))  
print(classification_report(y_test, outTree))  
```

```python

from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix  
from sklearn.model_selection import cross_val_score
from sklearn import tree

clf_gini = AdaBoostClassifier()
clrTree = clf_gini.fit(X_train, y_train)
outTree = clrTree.predict(X_test)

print(confusion_matrix(y_test, outTree))  
print(classification_report(y_test, outTree))  
```

```python
from sklearn.base import BaseEstimator

class AlwaysASBOClassifier(BaseEstimator):
    def fit(self, X, y=None):
        return self
    def predict(self, X):
        asbo = ['Anti-social behaviour'] * len(X)
        return np.array(asbo, dtype=object)
    
always_asbo_clf = AlwaysASBOClassifier()

clrTree = always_asbo_clf.fit(X_train, y_train)
outTree = clrTree.predict(X_test)

print(confusion_matrix(y_test, outTree))  
print(classification_report(y_test, outTree)) 
```
